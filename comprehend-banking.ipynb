```python
# Key phrases
phrases = comprehend.detect_key_phrases(Text=banking_txt, LanguageCode='en')

# Entities
entities = comprehend.detect_entities(Text=banking_txt, LanguageCode='en')

#Sentiments
sentiments = comprehend.detect_sentiment(Text=banking_txt, LanguageCode='en')
```
